```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
def answer(start, length):
    start = int(start)
    length = int(length)
    res = 0
    for i in range(length):
        for j in range(length-i):
            res=res^(start+j+length*i)
    return res
```

```python
def answer_arr(arrx, arry):
    l = len(arrx)
    new = np.empty(l)
    for i in range(l):
        z = answer(int(arrx[i]),int(arry[i]))
        np.append(new,z)
    return new
```

```python
def answerr(x,y):
    return answer(int(x),int(y))
```

```python
for x in range(10):
    for i in range(1,10):
        print()
        for j in [i*x for x in range(10)]:
            print (answer(x,j),end= ', ')
    print(' ')
```

```python
for i in range(10):
    print()
    for j in range(1,20):
        print (answer(0,j+i),end= ', ')
```

```python
# Data for three-dimensional scattered points
xdata = 1000 * np.random.random(100)
ydata = 1000 * np.random.random(100)
zdata = answerr(xdata,ydata)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(xdata, ydata, zdata, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
f = np.vectorize(answerr)

x = np.linspace(10, 20, 30)
y = np.linspace(10, 20, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```
